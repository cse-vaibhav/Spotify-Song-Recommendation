```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("top10s.csv", index_col=0)
dtypes = df.dtypes
numerical_cols = list(dtypes[dtypes != 'object'][1:].index)
categorical_cols = list(dtypes[dtypes == "object"].index)
df.shape
```

```python
df.year.unique()
```

```python
df.isnull().sum()
```

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10, 10))
sns.heatmap(corr, annot=True)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
sns.pairplot(df.drop(categorical_cols + ["year"], axis=1), diag_kind="kde")
plt.show()
```
